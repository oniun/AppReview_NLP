```python
import pandas as pd
import numpy as np
import re
```

### Data Import

```python
reviews = pd.read_csv('data/dataset.csv', encoding='utf8')
```

* Add Column ("review length")

```python
reviews['review_length'] = reviews['review'].apply(len)
```

### Preprocessing
* Remove some particles

```python
only_ko_en = list()

def lang_seperation(array):
    
    global only_ko_en 
    
    for item in array:
        text = re.compile('[^ ㄱ-ㅣ가-힣A-Za-z0-9♡\^~;♥❤❤️💓💘💕💞💟💜💛💖👍😍😘😻]+') 
        # hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')  # 위와 동일
        result = text.sub('', item) # 한글과 띄어쓰기를 제외한 모든 부분을 제거
        only_ko_en.append(result)

        rest = text.findall(item) # 정규식에 일치되는 부분을 리스트 형태로 저장
        if len(rest) >= 1:
            print(rest)
    
    return "Seperation Completed"
```

```python
rev = list(reviews['review'])
lang_seperation(rev)
reviews['review'] = only_ko_en
```

* app_name arrange

```python
applist = list(reviews['app'])
app_strip = list()
for app in applist:
    app_ = app.strip()
    app_strip.append(app_)
    
reviews['app'] = app_strip
```

* Negative/Postive Split

```python
# 1-4점은 부정의견, 5점은 긍정의견으로 분류
neg_rev = reviews[(reviews['rate'] == 1) | (reviews['rate'] == 2) | (reviews['rate'] == 3) | (reviews['rate'] == 4)]
pos_rev = reviews[reviews['rate'] == 5]
```

```python
neg_rev.tail()
```

```python
len(neg_rev)
```

```python
neg_rev['response'] = 'neg'
pos_rev['response'] = 'pos'
```

```python
from sklearn.utils import shuffle

df = pd.concat([neg_rev, pos_rev], axis=0)
df = shuffle(df)
df.tail()
```

* Making X, y data

```python
X = df['review']
y = df['response']
```

## Modeling

* Import Tokenizer(Twitter, Mecabm, kkma)

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import xgboost
from sklearn.pipeline import Pipeline
from nltk.corpus import stopwords
from konlpy.tag import Twitter
from konlpy.tag import Kkma
from konlpy.tag import Mecab

mecab = Mecab()
twitter = Twitter()
kkma = Kkma()

def mecab_tokenizer(text):
    return ['/'.join(t) for t in mecab.pos(text)]
def twit_tokenizer(text):
    return ['/'.join(t) for t in twitter.pos(text, norm=True, stem=True)]
def kkma_tokenizer(text):
    return ['/'.join(t) for t in kkma.pos(text)]
```

### Split Data(Train/Test)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2)
```

```python
%%time
pipe_1 = Pipeline([
    ('bow', CountVectorizer(tokenizer=mecab_tokenizer, min_df=3, ngram_range=(1, 3))),
    ('tfidf', TfidfTransformer()),
    ('model', LogisticRegression()),
])

pipe_1.fit(X_train, y_train)
prediction = pipe_1.predict(X_test)

print('accuracy: {}'.format(accuracy_score(y_test, prediction)))
```

```python
%%time
pipe_1 = Pipeline([
    ('bow', CountVectorizer(tokenizer=mecab_tokenizer, min_df=3, ngram_range=(1, 3))),
    ('tfidf', TfidfTransformer()),
    ('model', RandomForestClassifier(n_estimators=100, criterion='entropy', n_jobs=4)),
])

pipe_1.fit(X_train, y_train)
prediction = pipe_1.predict(X_test)

print('accuracy: {}'.format(accuracy_score(y_test, prediction)))
```

### stopword

```python
def set_stopword(array):
    stop_w = list()
    for item in array:
        for txt in (list(set(item))):
            if txt in stop_w:
                pass
            else:
                stop_w.append(txt)
    return stop_w

stopword_kp = [re.findall('[ㄱ-ㅎㅏㅑㅓㅕㅗㅛㅜㅠㅢㅐㅔㅒㅖ]+/KoreanParticle', str(twit_tokenizer(text))) for text in X]
stop_kp = set_stopword(stopword_kp)
len(stop_kp)

stopword_josa = [re.findall('[가-힣]+/Josa', str(twit_tokenizer(text))) for text in X]
stop_j = set_stopword(stopword_josa)

stopword_p = [re.findall('[.!?,''~^";]+/Punctuation', str(twit_tokenizer(text))) for text in X]
stop_p =set_stopword(stopword_p)


stop_kp.extend(stop_j)
stop_kp.extend(stop_p)

stopword = stop_kp
len(stopword)
```
